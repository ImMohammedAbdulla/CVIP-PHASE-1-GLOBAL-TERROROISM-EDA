```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("terrorism dataset.csv", encoding="latin1")
```

```python
dataset=dataset[["iyear","imonth","iday","country_txt","region_txt","provstate","city",\
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",\
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
dataset.shape
```

```python
dataset.head()
```

```python
dataset.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",\
                   "region_txt":"Region","provstate":"Province/State","city":"City",\
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",\
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",\
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",\
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
dataset.head()
```

```python
dataset.describe(include="all")
```

```python
dataset.isnull().sum()
```

```python
dataset["Killed"]=dataset["Killed"].fillna(0)
dataset["Wounded"]=dataset["Wounded"].fillna(0)
dataset["Damaged"]=dataset["Killed"]+dataset["Wounded"]
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset["Target Type"].unique()
```

```python
dataset["Target Type"].value_counts().plot(kind="bar",color="blue",figsize=(16,4))
plt.title("Types of Targets")
plt.xlabel("Target Types")
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks")
plt.show()
```

```python
dataset[["Target Type","Damaged"]].groupby("Target Type").sum().sort_values(by="Damaged").plot(kind="bar",color="blue",figsize=(16,4))
plt.title("People Damaged")
plt.xlabel("Target Types")
plt.ylabel("Number of People Damaged")
plt.show()
```

```python
dataset[["Target Type","Killed"]].groupby("Target Type").sum().sort_values(by="Killed").plot(kind="bar",color="blue",figsize=(16,4))
plt.title("People Killed")
plt.xlabel("Target Types")
plt.ylabel("Number of People Killed")
plt.show()
```

```python
dataset[["Target Type","Wounded"]].groupby("Target Type").sum().sort_values(by="Wounded").plot(kind="bar",color="blue",figsize=(16,4))
plt.title("People Wounded")
plt.xlabel("Target Types")
plt.ylabel("Number of People wounded")
plt.show()
```

```python
pd.crosstab(dataset.Year,dataset.Region).plot(kind="area",stacked=False,figsize=(16,12))
plt.title("Region wise attacks")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
region=pd.crosstab(dataset.Year,dataset.Region)
```

```python
regt=region.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(16,4))
plt.title("Total Number of Attacks in each Region")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```
